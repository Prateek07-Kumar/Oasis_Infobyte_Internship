**Data Cleaning and Preparation**

```python
car_data.head(1)
```

```python
# splitting company name from CarName columns

CompanyName = car_data['CarName'].apply(lambda x: x.split(' ')[0])
car_data.insert(3,"CompanyName",CompanyName)
car_data.drop(['CarName'],axis=1,inplace=True)
car_data.head(2)
```

```python
car_data.CompanyName.unique()
```

**There seems to be some spelling error in the CompanyName column.**

- maxda = mazda
- Nissan = nissan
- porsche = porcshce
- toyota = toyouta
- vokswagen = volkswagen = vw

```python
car_data.CompanyName = car_data.CompanyName.str.lower()

def replace_name(a,b):
    car_data.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

car_data.CompanyName.unique()
```

```python
# Checking for duplicates

car_data.loc[car_data.duplicated()]

# car_data.duplicated().sum()
```

**Visualizing the data**

```python
plt.Figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(car_data.price)



plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxenplot(y=car_data.price)

plt.show()
```

```python
print(car_data.price.describe(percentiles=[0.25,0.50,0.75,0.85,0.90,1]))
```

**Inference :**

1. The plot seemed to be right-skewed, meaning that the most prices in the dataset are low(Below 15,000).
2. There is a significant difference between the mean and the median of the price distribution.
3. The data points are far spread out from the mean, which indicates a high variance in the car prices.(85% of the prices are below 18,500, whereas the remaining 15% are between 18,500 and 45,400.)

```python
plt.Figure(figsize=(30,20))

plt.subplot(1,3,1)
plt1 = car_data.CompanyName.value_counts().plot(kind='bar',color='green')
plt.title('Companies Histogram')
plt1.set(xlabel = 'Car company', ylabel='Frequency of company')

plt.subplot(1,3,2)
plt1 = car_data.fueltype.value_counts().plot(kind='bar',color=['red','green'])
plt.title('Fuel Type Histogram')
plt1.set(xlabel = 'Fuel Type', ylabel='Frequency of fuel type')

plt.subplot(1,3,3)
plt1 = car_data.carbody.value_counts().plot(kind='bar',color=['yellow','green','blue','black','red'])
plt.title('Car Type Histogram')
plt1.set(xlabel = 'Car Type', ylabel='Frequency of Car type')

plt.show()
```

**Inference :**

1. Toyota seemed to be favored car company.
2. Number of gas fueled cars are more than diesel.
3. sedan is the top car type prefered.
